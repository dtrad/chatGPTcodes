```python
from fenics import *
import matplotlib.pyplot as plt
import numpy as np

# Create a 1D mesh of the interval [0, 1] with 32 elements
mesh = UnitIntervalMesh(32)

# Define function space (piecewise linear elements)
V = FunctionSpace(mesh, 'P', 1)

# Define boundary condition: u(0) = u(1) = 0
u_D = Constant(0.0)
def boundary(x, on_boundary):
    return on_boundary
bc = DirichletBC(V, u_D, boundary)

# Define source term f(x)
f = Expression('10*sin(pi*x[0])', degree=2)

# Define variational problem
u = TrialFunction(V)
v = TestFunction(V)
a = dot(grad(u), grad(v)) * dx
L = f * v * dx

# Compute solution
u = Function(V)
solve(a == L, u, bc)

# Plot solution
plot(u)
plt.title("Solution to Poisson's Equation with FEM")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.show()
```
