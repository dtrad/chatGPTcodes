```python
import numpy as np
import matplotlib.pyplot as plt

# Define two signals
t = np.linspace(0, 1, 100)
s1 = np.sin(2 * np.pi * 5 * t)
s2 = np.sin(2 * np.pi * 5 * t + np.pi / 4)

# Combine the signals
s_sum = s1 + s2

# Plot the signals
plt.plot(t, s1, label='Signal 1')
plt.plot(t, s2, label='Signal 2')
plt.plot(t, s_sum, label='Sum')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Define the Butterworth filter parameters
order = 2
cutoff_frequency = 20
nyquist_frequency = 0.5 * 100
normalized_cutoff = cutoff_frequency / nyquist_frequency

# Design the Butterworth filter
b, a = butter(order, normalized_cutoff, btype='highpass')

# Apply the filter to the seismic trace
filtered_trace = lfilter(b, a, trace)

# Plot the original trace and the filtered trace
plt.plot(trace, label='Original Trace')
plt.plot(filtered_trace, label='Filtered Trace')
plt.legend()
plt.show()
```
