```python
import numpy as np
import matplotlib.pyplot as plt

def ray_tracing(v, x0, theta0, t):
    x = x0 + v * t * np.cos(theta0)
    y = v * t * np.sin(theta0)
    return x, y

v = 1 # velocity of medium
x0 = 0 # initial x position
theta0 = np.pi / 4 # initial angle with respect to x-axis
t = np.linspace(0, 10, 100) # time values to compute ray path

x, y = ray_tracing(v, x0, theta0, t)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ray Tracing in a Constant Velocity Medium')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ray_tracing(v0, dvdx, x0, theta0, t):
    x = x0 + v0 * t * np.cos(theta0) + dvdx * (v0 * t * np.sin(theta0))**2 / 2
    y = v0 * t * np.sin(theta0) + dvdx * (v0 * t * np.cos(theta0))**2 / 2
    return x, y

v0 = 1 # initial velocity
dvdx = 0.1 # velocity gradient
x0 = 0 # initial x position
theta0 = np.pi / 4 # initial angle with respect to x-axis
t = np.linspace(0, 10, 100) # time values to compute ray path

x, y = ray_tracing(v0, dvdx, x0, theta0, t)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ray Tracing in a Medium with a Constant Velocity Gradient')
plt.show()
```
