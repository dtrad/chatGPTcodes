```python
# give a list of folders under /user/Daniel/Pictures/2024/ that contain images
root_folder = 'C:\\Users\\daniel.trad\\Pictures\\2024\\'
import os
import glob
for folder in glob.glob(root_folder + '\\202*'):
    if os.path.isdir(folder):
        if len(glob.glob(folder + '/*.*')) > 0:
            # copy contents of folder to root folder 
            # list files in folder
            for file in glob.glob(folder + '/*.*'):
                print('copying ' + file + ' to ' + root_folder)
                os.system('copy ' + file + ' ' + root_folder)
            
```
