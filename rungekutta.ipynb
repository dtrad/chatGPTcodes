```python
import numpy as np

def runge_kutta(ode_func, t0, y0, tf, dt):
    t = np.arange(t0, tf, dt)
    y = np.zeros((len(t), len(y0)))
    y[0,:] = y0

    for i in range(1, len(t)):
        k1 = dt * ode_func(t[i-1], y[i-1,:])
        k2 = dt * ode_func(t[i-1] + dt/2, y[i-1,:] + k1/2)
        k3 = dt * ode_func(t[i-1] + dt/2, y[i-1,:] + k2/2)
        k4 = dt * ode_func(t[i-1] + dt, y[i-1,:] + k3)
        y[i,:] = y[i-1,:] + (k1 + 2*k2 + 2*k3 + k4) / 6

    return t, y

def ode_func(t, y):
    # Example ODE: y' = -y, y(0) = 1
    return -y

t0 = 0
y0 = np.array([1.0])
tf = 10
dt = 0.01

t, y = runge_kutta(ode_func, t0, y0, tf, dt)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,y)
```
